# Import required python modules.

```python
import numpy as np
np.random.seed(7)
```

```python
from sklearn import datasets
```

# Load California Housing Price dataset.

```python
california_dataset = datasets.fetch_california_housing()
```

```python
print(california_dataset.keys())
```

```python
print(california_dataset.DESCR)
```

# Load the dataset.

### Import required python modules.

```python
import pandas as pd
```

```python
california = pd.DataFrame(
    california_dataset.data,
    columns=california_dataset.feature_names
)

california['MedHouseValue'] = california_dataset.target
california.head()
```

### Show dataset schema.

```python
california.info()
```

# Preprocess dataset.

### Check for missing values in the dataset.

```python
california.isnull().sum()
```

### Define input variables (X) and an output variable (y).

```python
X = california.drop('MedHouseValue', axis=1)
y = california['MedHouseValue']
```

```python
print(X.shape)
print(y.shape)
```

### Import required python modules.

```python
from sklearn.preprocessing import MinMaxScaler
```

### Compute scale factor and scaled the dataset.

```python
scaler = MinMaxScaler().fit(X)
scaled_X = scaler.transform(X)
```

### Import required python modules.

```python
from sklearn.model_selection import train_test_split
```

```python
seed = 9
test_size = 0.20
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size = test_size, random_state = seed)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```
